```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
# Load the dataset
df=pd.read_csv('./Churn_Modelling.csv')
```

```python
df.head()
# from the prevous input going to decide the person is going to exited or not
# this is past data
# here we are able to see first three columns not playing much role to predicting does person is leave the bank or not
# so first three columns we will delete
```

```python
df=df.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
df
```

```python
# convert age into 
# Encode categorical varibales
label_encoder_gender=LabelEncoder()
df['Gender']=label_encoder_gender.fit_transform(df['Gender'])
df
```

```python
# one hot encoding geography column
from sklearn.preprocessing import OneHotEncoder
one_ht_enc_geo=OneHotEncoder()
geo_encoder=one_ht_enc_geo.fit_transform(df[['Geography']])
geo_encoder
```

```python
geo_encoder.toarray()
```

```python
one_ht_enc_geo.get_feature_names_out(['Geography'])
# This are the three features 
```

```python
geo_encoder_df=pd.DataFrame(geo_encoder.toarray(),columns=one_ht_enc_geo.get_feature_names_out(['Geography']))
```

```python
geo_encoder_df
```

```python
# now concat data with original
df=pd.concat([df.drop('Geography',axis=1),geo_encoder_df],axis=1)
```

```python
df
```

```python
# save the encoder and scalar in pickle file
with open('label_encoder_gender','wb') as file:
    pickle.dump(label_encoder_gender,file)
with open('one_ht_enc_geo.pkl','wb') as file:
    pickle.dump(one_ht_enc_geo,file)
```

```python
# Divide the dataset into dependant and independent feature
# ex:-here exited is the column which dependent final output which is dependent on other remaining column we can say they are independent
```

```python
x=df.drop(['Exited'],axis=1) #x is independent
y=df['Exited'] # y is dependent 

# x → Features (independent variables)
# y → Target (dependent variable)
# test_size=0.2 → 20% of the data is allocated for testing, and 80% for training.
# random_state=42 → Ensures the split is reproducible (so you get the same split every time you run the code).

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train
# now scale this features
# normally apply it on only independent like x axis
scalar=StandardScaler()
# for converting values in between 0 to 1 
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
x_train




```

```python
# now save this in pkl file
with open('scalar.pkl','wb') as file:
    pickle.dump(scalar,file)
```
